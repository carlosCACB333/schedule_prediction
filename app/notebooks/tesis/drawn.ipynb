```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sigmoide Function


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 100)
y = sigmoid(x)
sns.lineplot(x=x, y=y, label="Funci√≥n Sigmoide")
plt.xlabel("x")
plt.ylabel("y")
plt.text(2, 0.5, r"$f(x) = \frac{1}{1 + e^{-x}}$", fontsize=18, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
# tanh Function


def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))


x = np.linspace(-10, 10, 100)
y = tanh(x)

sns.lineplot(x=x, y=y, label="Funci√≥n Tangente Hiperb√≥lica")
plt.xlabel("x")
plt.ylabel("y")
plt.text(2, 0.5, r"$f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$", fontsize=18, fontweight="bold")
plt.grid(True)
plt.show()

```

```python
# ReLU Function


def relu(x):
    return np.maximum(0, x)


x = np.linspace(-10, 10, 100)
y = relu(x)

sns.lineplot(x=x, y=y, label="Funci√≥n ReLU")
plt.xlabel("x")
plt.ylabel("y")
plt.text(2, 0.5, r"$f(x) = max(0, x)$", fontsize=18, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
# grafic Binary Cross-Entropy

fig, ax = plt.subplots(1,2, figsize=(15,5))

x = np.linspace(0.0001, 1, 100)
y = -np.log(x)
sns.lineplot(x=x, y=y, label="Funci√≥n de Costo Binaria", ax=ax[0])
ax[0].set_xlabel("y")
ax[0].set_title("y = 1")
ax[0].set_ylabel("J(y)")
ax[0].text(0.5, 2, r"$J(y) = -log(y)$", fontsize=18, fontweight="bold")
ax[0].grid(True)

x = np.linspace(0.0001, 1, 100)
y = -np.log(1-x)
sns.lineplot(x=x, y=y, label="Funci√≥n de Costo Binaria", ax=ax[1])
ax[1].set_xlabel("y")
ax[1].set_title("y = 0")
ax[1].set_ylabel("J(y)")
ax[1].text(0.5, 2, r"$J(y) = -log(1-y)$", fontsize=18, fontweight="bold")
ax[1].grid(True)

plt.show()


```
